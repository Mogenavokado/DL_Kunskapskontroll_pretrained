```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

# Load Pretrained model(ResNet50) and database(ImageNet) 

```python
model = ResNet50(weights='imagenet')
# ResNet stands for Residual Network and is a specific type of convolutional neural network (CNN).
# ResNet-50 is a 50-layer convolutional neural network (48 convolutional layers, one MaxPool layer, and one average pool layer).
# Imagenet is a large database used for object identification.
```

```python
#img_path = 'zebra.jpg'
# img_path = 'linjal.jpg'
# img_path = 'tennisboll.jpg'
img_path = 'säng.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img
```

# Preprocess step

```python
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x, axis=0)
x.shape
```

```python
# https://www.tensorflow.org/api_docs/python/tf/keras/applications/resnet50/preprocess_input
x = preprocess_input(x)
```

# Make predictions

```python
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=2)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```

# 2. Answer the question: Why is "Pretrained models" a very powerful concept?

Pretrained models is a very powerful concept in manners that it already has knowledge learned from large-scale datasets and already been trained. This makes it easier and faster to develop models for new tasks or datasets i.e from a large-scaled pretrained model we can fine-tuned it on a smaller dataset for a specific task. Since it already been trained and has a foundational knowledge, the model can quickly adapt to new tasks.
